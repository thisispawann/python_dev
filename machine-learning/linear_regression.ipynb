```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("homeprice.csv")
df
```

```python
%matplotlib inline

plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color='green', marker='*')
```

```python
new_df = df.drop('price', axis='columns')
new_df
```

```python
# creating object of Linear Regression model

model = linear_model.LinearRegression()
model
```

```python
# train the model
model.fit(new_df, df.price)
```

```python
# Linear Regression takes two dimensional array to predict the model
model.predict([[3300]])
```

```python
# Explanation on how predict function actually works:

# As we know that the equation: y = mx + c where m is slope or gradient and c is intercept
model.coef_
```

```python
model.intercept_
```

```python
# y = mx + c now lets apply their value
135.78767123*3300 + 180616.43835616432
```

```python
# so we have same value: 628715.7534151643
```

```python
# let suppose we have a areas data now we have to predict their prices with this model
# let's import dataset
area_df = pd.read_csv("areas.csv")
area_df
```

```python
# now predict the prices
price = model.predict(area_df)
price
```

```python
#now add price column and insert their values
area_df['prices'] = price
area_df
```

```python
# save into the csv format
area_df.to_csv("predicted_areas.csv")
```
